```python
data = {
    "text": [
        "Congratulations! You won a free lottery, claim now!",
        "Urgent: Update your bank details to avoid account suspension.",
        "Hi, let's catch up for coffee tomorrow.",
        "You have been selected for a prize! Click here to claim.",
        "Meeting at 3 PM, please be on time."
    ],
    "label": ["spam", "spam", "ham", "spam", "ham"]
}

df = pd.DataFrame(data)
```

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    return text
```

```python
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df["text"])
y = df["label"]

model = MultinomialNB()
model.fit(X, y)
```

```python
import tldextract

def detect_phishing(text):
    # Extract URLs from the text
    urls = re.findall(r"http[s]?://[^\s]+", text)

    phishing_domains = [
        "bank-verification", "secure-login", "account-update",
        "paypal-security", "confirm-payment", "verify-now"
    ]

    for url in urls:
        extracted = tldextract.extract(url)
        domain = extracted.domain.lower()

        for keyword in phishing_domains:
            if keyword in domain:
                return True  # Phishing detected
    
    return False  # No phishing links found
```
